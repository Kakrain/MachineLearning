```python
def get_cicle_data(n=50, r=1):
    
    np.random.seed(n)
    
    thetas = np.random.uniform(0, 2*np.pi, n)
    data = np.array([r * np.sin(thetas), r * np.cos(thetas)]).T + np.random.normal(scale=1/8, size=(n,2))
    
    return data
```

```python
def get_gaussian_data(n, mean=0.0, sigma=1.0):
    
    np.random.seed(n)
    
    data = np.random.normal(loc=mean, scale=sigma, size=(n,2))
    
    return data
```

```python
def PCA(xs, l):
    n=len(xs)
    d=len(xs[0])
    m=np.mean(xs)
    xm=np.add(xs,-m)
    C=np.matmul(np.transpose(xm),xm)
    w,v=np.linalg.eig(C)
    sortind=np.argsort(w)
    wl=np.sort(w)[l:]
    vl=np.transpose(np.delete(np.transpose(v),sortind[:l],axis=0))
    xl=np.transpose(np.delete(np.transpose(xm),sortind[:l],axis=0))
    y=np.matmul(vl,np.transpose(xl))
    P=np.matmul(vl,np.transpose(vl))
    z=np.transpose(np.matmul(P,np.transpose(xm))+m)
    return (y,z,np.flip(np.sort(w),axis=0),wl)

```
