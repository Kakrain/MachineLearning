## Exercise 1

```python
def get_cicle_data(n=50, r=1):
    
    np.random.seed(n)
    
    thetas = np.random.uniform(0, 2*np.pi, n)
    data = np.array([r * np.sin(thetas), r * np.cos(thetas)]).T + np.random.normal(scale=1/8, size=(n,2))
    
    return data
```

```python
def get_gaussian_data(n, mean=0.0, sigma=1.0):
    
    np.random.seed(n)
    
    data = np.random.normal(loc=mean, scale=sigma, size=(n,2))
    
    return data
```

### b)

```python
def PCA(xs, l):
    m=np.mean(xs)
    xm=xs-m
    C=np.matmul(xm.T,xm)
    w,v=np.linalg.eig(C)
    sortind=np.argsort(w)
    wl=np.sort(w)[l:]
    vl=np.delete(v.T,sortind[:l],axis=0).T
    xl=np.delete(xm.T,sortind[:l],axis=0).T
    y=np.matmul(np.transpose(vl),np.transpose(xm))
    P=np.matmul(vl,np.transpose(vl))
    z=np.transpose(np.matmul(P,np.transpose(xm))+m)
    return (y,z,np.flip(np.sort(w),axis=0),wl)

y,z,w,wl=PCA(data,1)
```

### b)

```python
def kernel_PCA(kernel, l=2):
    n = len(kernel)
    onen = np.zeros((n,n)) 
    K = kernel - np.matmul(onen, kernel) - np.matmul(kernel, onen) + np.matmul(np.matmul(onen, kernel), onen)
    u, s, vh = np.linalg.svd(K)
    Vl = np.zeros((n,l))
    for i in range(l):        
        Vl[:,i] = u[:,i]/np.sqrt(s[i])
    
    y = np.zeros((n,l))
    for i in range(l):
        for j in range(n):
            y[j,i] = y[j,i] + np.multiply(Vl[j,i], K[j,i])
    
    return y
```
